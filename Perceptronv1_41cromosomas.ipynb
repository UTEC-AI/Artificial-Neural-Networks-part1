# Perceptron implementation

## Gradient descent
![gd](https://cdn-images-1.medium.com/max/1600/1*f9a162GhpMbiTVTAua_lLQ.png)

```python
class Perceptron:
    def __init__(self, inputdim=2):
        self.input_dim = inputdim
        self.W = np.random.normal(size=(1, self.input_dim))
        self.b = np.random.normal()
    
    def f(self, x):
        return 1./(1. + np.exp(-x))

    
    def predict(self, input_data):
        sum_w = np.dot(input_data, self.W.T)
        u = sum_w + self.b
        y = self.f(u)
        return y
        
        
    def train(self, inputs, outputs, alpha=0.01, epochs=300):
        errors = []
        for i in range(epochs):
            X = inputs
            Ye = outputs
            Y = self.predict(X)
            dW = -alpha*np.dot(((Ye-Y)*Y*(1-Y)).T, X)
            dB = alpha*np.dot(((Ye-Y)*Y*(1-Y)).T, np.ones((4, 1)))
            self.W += dW
            self.b += dB
            errors.append(np.mean(np.square(Ye-Y)))
        plt.plot(errors)
```

```python
AND_X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
AND_Y = np.array([[0], [0], [0], [1]])
```

```python
p = Perceptron(inputdim=2)
p.train(AND_X, AND_Y, alpha=0.09, epochs=100)
```
